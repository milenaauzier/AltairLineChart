```python
import panel as pn 
import altair as alt
from altair import datum 
import pandas as pd
from vega_datasets import data 
import datetime as dt 
```

```python
alt.renderers.enable('default')
pn.extension('vega')
```

```python
source = data.stocks()
```

```python
source.head()
```

```python
source.rename(columns={'symbol': 'Empresa'}, inplace = True)
source.rename(columns={'price': 'preço'}, inplace = True)
source.rename(columns={'date': 'Data'}, inplace = True)
```

```python
source.head()
```

```python
lines = (
        alt.Chart(source)
        .mark_line()
        .encode(x="Data", y = "preço", color = "Empresa").properties(title="Preço das Ações")
).interactive()
```

```python
xrule = (
        alt.Chart()
        .mark_rule(color = "cyan", strokeWidth = 2)
        .encode(x = alt.datum(alt.DateTime(year = 2008, month = "November")))
) 
```

```python
yrule = (
      alt.Chart().mark_rule(strokeDash=[12,6], size = 2).encode(y=alt.datum(350)))
```

```python
lines + yrule + xrule 
```
